# KÖK BULMA

[a,b]$\in$x aralığında tanımlı bir f(x) fonksiyonunu sıfır yapan x değerini bulma işlemi;  
o fonksiyonun köklerini bulmak olarak adlandırılır.

Diğer bir ifadeyle $f(x) = 0$ eşitliğini sağlayan x değerleri aranılan köklerdir.

## YARILAMA YÖNTEMİ (INTERVAL / BISECTION METHOD)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

isaretle = lambda x, y: plt.plot([0,x,x],[y, y,0], '--')

a = 5.0
b = 8.5
xo = (a+b)/2.

x = np.linspace(a, b, 100)
f = lambda x: x*np.sin(x)

isaretle(a, f(a))
isaretle(b, f(b))
isaretle(xo, f(xo))

plt.plot(x,f(x), color="red")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(color='black')
plt.xlim(a*.9,b*1.1)
plt.show()
```

```python
a0 = 5.0
b0 = 8.5
tol = 1e-6

a = a0
b = b0
dkok = abs(b-a)

f = lambda x: x*np.sin(x)


while dkok>tol:
    xo = (a+b)/2.
    if f(b)*f(xo)>0:
        b = xo
    else:
        a=xo
    dkok = abs(b-a)

print 'aranan kök = %s'%xo
```

```python
a0 = 5.0
b0 = 8.5
tol = 1e-6

a = a0
b = b0
dkok = abs(b-a)
x = np.linspace(a, b, 100)
f = lambda x: x*np.sin(x)

plt.figure(figsize=(9,9))
plt.plot(x,f(x), color="red")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(alpha=.5)
plt.axhline(color='black')
plt.xlim(a0, b0)

it = 0
while dkok>tol:
    xo = (a+b)/2.
    if f(b)*f(xo)>0:
        b = xo
    else:
        a=xo
    # a ve b bir birine ne kadar yakın
    dkok = abs(b-a)
    # döngü sayısı
    it += 1
    fxo = f(xo)
    isaretle(xo, fxo)
    plt.plot(xo, fxo,'o')
    plt.text(xo,fxo, str(it), fontsize=16)

print "%s denemede aşağıdaki kök bulundu"%it
print 'aranan kök = %s'%xo
print 'f(%s)=%s'%(xo, f(xo))
plt.show()
```

## KİRİŞ YÖNTEMİ (SECANT METHOD)

```python
def kiris(x0,x1, tol, itermax=50):
    dx = x1-x0
    iter = 1
    while abs(dx)>tol:
        x2 = x1 - (x1-x0)*f(x1)/(f(x1)-f(x0))
        x0, x1 = x1, x2
        dx = x1 - x0
        if iter>itermax:
            print "%s denemede kök bulunamadı"%itermax
            break
        iter += 1
    return x2
```

```python
def kiris_kokListesi(x0,x1, tol, itermax=50):
    dx = x1-x0
    xkok = [x0, x1]
    fxkok = [f(x0), f(x1)]
    iter = 1
    while abs(dx)>tol:
        x2 = x1 - (x1-x0)*f(x1)/(f(x1)-f(x0))
        xkok += [x2]
        fxkok += [f(x2)]
        x0, x1 = x1, x2
        dx = x1 - x0
        if iter>itermax:
            print "%s denemede kök bulunamadı"%itermax
            break
        iter += 1
    return x2, xkok, fxkok
```

## NEWTON-RAPHSON YÖNTEMİ

```python
# ilk yazdığımız newton_raphson kök bulma fonksiyonu
def newton_eski(x1, tol):
    x2 = 3.*tol
    dkok = abs(x2-x1)
    iter=1
    while dkok>tol:
        x1 = x2
        x2 = x1 - f(x1)/f1(x1)
        dkok = abs(x2-x1)
        iter += 1
        if iter>5:
            print ("kök bulunamadı")
            break
    return x2
```

```python
# yeniden düzenlenmiş olan
def newton(x2, tol=1.0e-8, itermax=5):
    iter=1
    while iter<itermax:
        x1 = x2
        x2 = x1 - f(x1)/f1(x1)
        dkok = abs(x2-x1)
        iter += 1
        if dkok<=tol: break
    if dkok>tol:
        print "%s denemede, aranan köke %s kadar yaklaşılabilmiştir."%(iter, dkok)
        print "İstenilen hassasiyet için itermax veya x başlangıç değeri değiştirilmelidir."
    return x2
```

## Karacisim Işıması

$\LARGE \bf u(\lambda, T) = \frac{8\pi h c}{\lambda^5}\frac{1}{\text{e}^{hc/k_BT\lambda}-1}$ denklemi için $\large \bf x = hc/k_BT\lambda$ değişken dönüşümü yapılırsa;

$\Large \bf u(x) = A \frac{x^5}{\text{e}^x-1}$ elde edilir. Burada A gerekli dönüşüm yapıldıktan sonra geriye kalan sabitleri temsil etmektedir.

Türevinin sıfır olduğu x değeri bulunarak $\large \bf \lambda_{max}$ bulunur.

```python
import numpy as np
import sympy as sm
import matplotlib.pyplot as plt
%matplotlib inline

x = sm.symbols('x') 
f = lambda x: (5. - x) - 5.*sm.exp(-x)
fn = sm.lambdify(x, f(x), "numpy")

a = 0.0
b = 10.0
x = np.linspace(a, b, 100)
y = fn(x)

plt.figure(figsize=(9,9))
plt.plot(x, y, color="blue", label="f(x)")
plt.axhline(color='black')
plt.grid()
plt.legend()
plt.show()
```
